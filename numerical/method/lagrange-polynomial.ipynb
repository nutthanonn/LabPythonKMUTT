```python
import numpy as np
import matplotlib.pyplot as plt


x_axis = [0, 1, 3/4]
y_axis = [2, 0, .25]

n = 3


def Lagrange(x):
    global n
    res = 0
    for i in range(n):
        l_mul = 1
        for j in range(n):
            if j != i:
                l_mul *= (x - x_axis[j]) / (x_axis[i] - x_axis[j])
        
        res += l_mul * y_axis[i]
    return res
```

```python
x_plot = np.linspace(1, 1000, 100000)
y_plot = Lagrange(x_plot)
plt.plot(x_plot, y_plot)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def secant(x0, x1, f, iteration=20):

    i = 0
    while np.sign(f(x0) - f(x1)) < 1E-10 and i < iteration:
        x = x1 - (( x0 - x1) / (f(x0) - f(x1)))*f(x1)
        x0 = x1
        x1 = x
        print(f"Iteration {i} - x = {x} - [ f(x0) - f(x1) ] = {f(x0) - f(x1)}, f(x1) = {f(x1)}")
        i += 1
    return x


def ShowGraph(x, f):
    x_plot = np.linspace(x-2, x+2, 1000)
    y_plot = f(x_plot)
    fig = plt.figure()
    plt.plot(x_plot, y_plot, c='blue')
    plt.plot(x, f(x), c='red', marker='o', fillstyle='none')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
    plt.show()
```

```python
def f(n): return 25000*(1 + 12/100)**n - 250_000

x = secant(15, 19, f)
print(f"result = {x:.4f}")
ShowGraph(x, f)
```

```python
def f(x): return x**3 - 5*x + 1

x = secant(0, 1, f)
print(f"result = {x:.4f}")
ShowGraph(x, f)
```

```python
def f(x): return x**3 - 3*x + 2

x = secant(5, 10, f, iteration=30)
print(f"result = {x:.4f}")
ShowGraph(x, f)
```

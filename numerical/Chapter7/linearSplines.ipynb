```python
import numpy as np
import matplotlib.pyplot as plt

def LinearSpline(x, xi, yi):
    nx = len(xi)
    ny = len(yi)

    assert nx == ny, "xi and yi must have the same length"
    assert sorted(xi) == list(xi), "xi must be sorted"

    iplus1 = [ind for ind,val in enumerate(xi) if val>x][0]
    i = iplus1 - 1

    return yi[i] + (yi[iplus1] - yi[i])*(x - xi[i])/(xi[iplus1] - xi[i])
```

```python
a=(-4,1,2,3,5)
fa=(1182,2,48,272,2262)
n=np.arange(-4,5,0.0010)
y=[LinearSpline(i,a,fa) for i in n]

plt.plot(n,y)
plt.plot(a,fa,'ro')
```
